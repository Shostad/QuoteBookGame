```python
text = open("quotes.txt",'r')
# out = open("output",'w')
```

```python
textRaw = ""
for x in text:
    textRaw += x
print(textRaw)

textSplit = textRaw.split("\n\n")
# print(textSplit)
```

```python
text = []
people = []
date = []
currentQuote = []
for x in textSplit:
    currentQuote = x.split('-')
    # print(currentQuote)
    text.append(currentQuote[0])
    namesAndDate = currentQuote[1].split(' ')
    date.append(namesAndDate[-1])
    people.append(namesAndDate[0:-1])

# getting rid of empty strings
for x in people:
    for y in range(len(x)-1):
        if x[y] == '':
            x.remove(x[y])
# print(currentQuote)
print(people)
```

```python
# formatting as sql inserts
names = []
uniqueNames = []
for x in people:
    for y in x:
        names.append(y)
for x in names:
    if x not in uniqueNames:
        uniqueNames.append(x)
print(uniqueNames)
textOut = "insert into person (name)"
```
